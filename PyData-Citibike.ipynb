If we want to estimate the number of trips, we only want one side of the diffferences.  So we want to grab only the rows where the difference is negative.  Summing these should give us the total number of trips

```python
diff_df[PARK_47th].plot()
plt.figure()
diff_df[diff_df[PARK_47th] > 0][PARK_47th].plot()
```

```python
trip_starts = diff_df[diff_df > 0]
#don't do this, you are poluting the global namespace with new variables,  
# I frequently do it, but regret it.  Having all these vraiable names strewn about
# makes autocomplete more difficult to use
#P47_diff = diff_df[PARK_47th]
#P47_starts = P47_diff[P47_diff > 0]

#instead write a function like this
def station_trip_starts(station_id):
    station_ts = diff_df[station_id]
    station_starts = station_ts[station_ts > 0]
    return station_starts
trip_starts[PARK_47th][D1:D2].cumsum().plot()
#refine this plot
# it wouldn't have a lot of nulls if it was a plot of only that station 
# where diff_df > 0, instead it includes any time period where other stations
# may have had a trip start
plt.figure()
station_trip_starts(PARK_47th)[D1:D2].cumsum().plot()
#notice how the second plot is smooth?
```

The cumsum function returns the cumulative sum. plotting it gives a running view of the total number of trips.


dataframe.sum() will return a dataframe with as many columns as the original, with the result of each column being the sum of that column over the time period.   This will be a dataframe for us with 330 columns and 1 row

dataframe.sum(axis=1) will return a dataframe with a single column and as many rows as the original.  It will be the timewise sum of started trips.

Taking averages
here we will line everything up to the same day and see the average station population by time of day

Time of day based population averages
-------------------------------------



```python
def make_baseline_df(df, time_period="1min", aggregate="mean"):
    """ find the average population over all dates in df, compress it to a single day"""
    df2 = df.copy()
    df2.index = df2.index.hour * 3600 + df2.index.minute *60 + df2.index.second + rand()
    df2 = df2.sort_index()
    df2.index = df2.index.map(lambda x: dtdt(2013,3,1) + dt.timedelta(0, int(x), rand()))
    df4 = df2.resample(time_period, aggregate)
    return df4
```

```python
total_avg = make_baseline_df(two_station_pop)
total_avg.Park_47th.plot()
```

I'm frequently asked "can you tell me when this dock will be full so I don't get stuck searching for an empty dock".  With the average population we can answer on average when a dock will be full and when it will be empty.  This is not a great answer, any New Yorker could tell you that midtown will be full all day.  A more interesting question is "How long at a given station should it take before a dock opens up at a given time of day".  Even though midtown is full in the middle of the day, people do still ride from midtown to other parts of the city.  How frequently does that happen?  There is enough data here to figure that out.

Having a single day average is useful, but it is hard to do comparisons against.  I have written functions which take that average and splay it over every time period in the data range.


```python
def extend_df(day_df, full_index):
    """take an average that occurs on a single day and stretch that average over all days  """
    if (day_df.index[-1] - day_df.index[0]) > ONE_DAY:
        raise Exception("extend_df only works for a day_df with a size of less than a day")
    days = np.unique(full_index.map(lambda x: dtdt(x.year, x.month, x.day)))
    day_dfs = []
    for d in days:
        day_df2 = day_df.copy()
        day_df2.index = day_df2.index.map(lambda x: dtdt(d.year,d.month, d.day, x.hour, x.minute,x.second))
        day_dfs.append(day_df2)
    full_df =  pd.concat(day_dfs)
    
    return full_df.sort_index()


def make_full(base_df):
    base_df['weekday'] = base_df.index.map(lambda x: x.weekday())
    weekday_df = base_df[base_df['weekday'] < 5]
    weekend_df = base_df[base_df['weekday'] >= 5]

    wDay_single_day_df = make_baseline_df(weekday_df)
    wEnd_single_day_df = make_baseline_df(weekend_df)
    wDay_full_df = extend_df(wDay_single_day_df,weekday_df.index)
    wEnd_full_df = extend_df(wEnd_single_day_df,weekend_df.index)
    base_df.index = base_df.index.map(lambda x: dtdt(x.year,x.month, x.day, x.hour, x.minute))
    full_df = pd.concat([wDay_full_df, wEnd_full_df]).sort_index()
    del wDay_full_df
    del wEnd_full_df
    full_df.index = full_df.index.map(lambda x: dtdt(x.year,x.month, x.day, x.hour, x.minute))
    """aligns the two dataframes so that they share every index, 
       filled with last value carried forward"""
    base_df, full_df = base_df.align(full_df, join="outer", axis=0, method="ffill", copy=False)
    return base_df, full_df
```

```python
two_station_pop, full_avg_df = make_full(two_station_pop)
```

```python
df, full_preDeriv_df = make_full(two_station_pop.diff(DIFFERENTIAL_TIMEPERIOD))
def find_changes(was_rebalanced_df):
    """ this finds all places where the was_rebalanced_df changes from False, to True  """
    rebalance_count_df = pd.DataFrame(index=was_rebalanced_df.index)
    for col in df.columns:
        #if not type(col) == int:
        #    continue
        convolved = np.convolve(was_rebalanced_df[col], [-1, 1])
        places_where_changed = convolved == 1
        
        rebalance_count_df[col] = places_where_changed[0:-1]
    return rebalance_count_df

def make_rebalance_count(avg_df, actual_df, threshhold, comp="<", *extra_disqaulifiers):
    diff_df = avg_df - actual_df
    if comp == "<":
        filtered = diff_df < threshhold
    else:
        filtered = diff_df > threshhold
    for dq_df in extra_disqaulifiers:
        filtered[dq_df == False] = False
    return diff_df, find_changes(filtered)

pdiff, pop_rebalance = make_rebalance_count(full_avg_df, two_station_pop, -10)
pop_rebalance.Park_47th.resample('D', how='sum').plot()
```

```python

def show_rebalancings(station_num, rebalancing_df, orig_df, baseline_df, window_width=4, outer_window_width=8):
    """ Rebalancing_df is teh boolean dataframe where True means a rebalancing took place"""
    station_column = rebalancing_df[station_num]
    #this is a much smaller time series with entries only for times
    # when rebalancing took place
    rebalance_points = station_column[station_column==1]
    
    for t in rebalance_points.index:
        t1, t2 = [t-(window_width*ONE_HOUR),t+(window_width*ONE_HOUR)]
        ot1, ot2 = [t-(outer_window_width*ONE_HOUR),t+(outer_window_width*ONE_HOUR)]
        base_ts = baseline_df[ot1:ot2][station_num]
        actual_ts = orig_df[t1:t2][station_num]
        yield [base_ts, actual_ts, t]
def plot_rebalancing(ts1, ts2, t):
    ts2.plot(legend=True, )
    plt.title("rebalancing at %r" %t)
    ts1.plot(label="weekeday_average")
    
```

```python
sample_size = '15min'
minimum_negative_average_slope = -.2
diff_parameter = -1
begin_time, end_time = [two_station_pop.index[0], two_station_pop.index[-1]]
#begin_time, end_time = [D1,D3]
full_avg_post_deriv = full_avg_df.diff(DIFFERENTIAL_TIMEPERIOD)
resamp_avg_deriv = full_avg_post_deriv[begin_time:end_time].resample(sample_size)
resamp_act_deriv = df.diff(DIFFERENTIAL_TIMEPERIOD)[begin_time:end_time].resample(sample_size)
deriv_diff,  deriv_rebalance = make_rebalance_count(
        resamp_avg_deriv, resamp_act_deriv, diff_parameter,  '<',
        resamp_avg_deriv < minimum_negative_average_slope, resamp_act_deriv > 0.001)
plt.figure(figsize=(28, 5), dpi=100)
deriv_rebalance.sum(axis=1).resample('1W', how='sum').plot()                                                               
print len(deriv_rebalance[deriv_rebalance.Park_47th == True].Park_47th)
```

```python

rb_iter = show_rebalancings('Park_47th', deriv_rebalance, two_station_pop, full_avg_df, window_width=2)
i = 0
plt.figure(figsize=(28, 5), dpi=100)
for param in rb_iter:
    plot_num = (i%3) + 1
    plt.subplot(1,3,plot_num)
    plot_rebalancing(*param)
    i+=1
    if plot_num ==3:
        plt.figure(figsize=(28, 5), dpi=100)
```
